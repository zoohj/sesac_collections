종로구의 미세먼지 농도, 초미세먼지 농도의 1달치 데이터를 정리하시오. (pm10Value,pm25Value)

```python
import requests

url = 'http://apis.data.go.kr/B552584/ArpltnInforInqireSvc/getMsrstnAcctoRltmMesureDnsty'
params ={'serviceKey' : 'tz/5tBf2Y05yFA2M9go/iqk9gIwYTvUgLGg+9KDyASXRyyiZSDlKvwOECcPro8oBlZwrEhzpYJPNRlxBY3VKSg==', 
         'returnType' : 'json', 
         'numOfRows' : '100', 
         'pageNo' : '1', 
         'stationName' : '종로구', 
         'dataTerm' : 'MONTH',
         'ver' : '1.0' }
try:
    response = requests.get(url,params=params)
    response.raise_for_status()
    data = response.json()
    lsts = data['response']['body']['items']
    answer=[]
    for lst in lsts:
        dict={'dataTime':lst['dataTime'], 'pm10Value':lst['pm10Value'],'pm25Value':lst['pm25Value']}
        answer.append(dict)
    pprint("종로구 한달간 미세먼지농도 및 초미세먼지 농도")
    print()
    pprint(answer)

except Exception as e:
    print(e)
```
