# 자료형(Data types)

## 1. 자료형이란?
자료형(data types)은 프로그래밍을 할 때 사용되는 모든 자료(데이터)의 형태를 의미합니다. 어떤 형태의 자료를 처리할 것인지 알아야만 올바른 결과물을 만들 수 있습니다.

---
## 2. 숫자형

### 2-1. 정수형(int)
양의 정수, 음의 정수, 0을 뜻합니다.

```python
print(100)
print(0)
print(-25)

print(type(100))
print(type(0))
print(type(-25))
```

### 2-2. 실수형(float)
소수점이 포함된 숫자를 뜻합니다.

```python
print(1.2)
print(-3.56)

print(type(1.2))
print(type(-3.56))
print(type(4.0))  # 정수 4와 같아 보이지만 소수점이 있으므로 실수형입니다.
```

---
## 3. 문자열(String)

### 3-1. 문자열이란?
한글, 알파벳 등 문자들의 집합을 뜻하며, 작은따옴표(') 혹은 큰따옴표(")로 감싸 표기합니다.

```python
print('Hello Python')
print('파이썬')
print('123')    
print('#$%^&*')

print(type('python'))
print(type("python"))
```

### 3-2. 문자열 사용 시 주의사항

1. **공백과 빈 문자열**: 공백(`' '`)과 빈 문자열(`''`)도 하나의 문자열로 취급됩니다.
2. **따옴표 짝 맞추기**: `'Hello"`와 같이 섞어 쓰면 `SyntaxError`가 발생합니다.
3. **따옴표 중첩**: 같은 종류의 따옴표를 중첩하면 에러가 발생합니다.

```python
print("큰따옴표 안에 '작은따옴표' 사용하기")
print('작은따옴표 안에 "큰따옴표" 사용하기')

# 에러 예시 (주석을 풀고 실행해 보세요)
# print("큰따옴표 안에 "큰따옴표" 사용하기")
```

### 이스케이프 시퀀스(Escape Sequence)
백슬래시(`\`)를 사용하여 특수한 문자를 표현합니다.

| 시퀀스 | 설명 | 예시 |
|:---:|:---|:---|
| `\'` | 작은따옴표 표현 | `'\'이렇게\''` |
| `\"` | 큰따옴표 표현 | `"\"이렇게\""` |
| `\\` | 백슬래시 표현 | `"\\"` |
| `\n` | 줄바꿈(Next line) | `"Hello\nPython"` |
| `\t` | 탭(Tab) | `"Name:\tKyle"` |

```python
print("큰따옴표 안에 \"이렇게\" 사용하기")
print('작은따옴표 안에 \'이렇게\' 사용하기')
print("하나\n둘\t셋")
```

### ⚠️ 숫자 123과 문자열 '123'은 다릅니다.
외관상 같아 보여도 자료형에 따라 할 수 있는 연산이 달라집니다.

```python
print(type(123))
print(type('123'))
```

## 4. 불린형(Boolean)
참(`True`)과 거짓(`False`)을 뜻합니다.

```python
print(True)
print(False)
print(type(True))
```

---
## 5. 명시적 형변환(Typecasting)

### 5-1. `int()`: 정수형으로 변환
- 문자열(정수 형태) → 정수
- 실수형 → 정수 (소수점 이하 버림)

```python
print(int('123'))
print(int(1.5))   # 1
print(int(-1.5))  # -1

# 에러 주의: 실수 형식의 문자열은 int()로 바로 변환 불가
# print(int('1.5'))  # ValueError 발생
```

### 5-2. `float()`: 실수형으로 변환
- 문자열(실수 형태) → 실수
- 정수형 → 실수

```python
print(float('3.5'))
print(float(10))   # 10.0

# 에러 주의: 계산식 형태의 문자열은 변환 불가
# print(float('3/5')) # ValueError 발생
```

### 5-3. `str()`: 문자열로 변환

```python
print(str(123))
print(str(10.5))
print(str(True))
print(type(str(True)))
```

### 5-4. `bool()`: 불린형으로 변환
파이썬에서는 데이터의 유무에 따라 참과 거짓을 평가합니다.

| 데이터의 값 | Boolean | 설명 |
|:---|:---:|:---|
| `0`, `0.0` | **False** | 숫자 0은 거짓 |
| `''` (빈 문자열) | **False** | 내용이 없으면 거짓 |
| `[]` (빈 리스트) | **False** | 비어있는 리스트는 거짓 |
| `1`, `-1`, `1.5` | **True** | 0 이외의 숫자는 참 |
| `'Hello'` | **True** | 내용이 있으면 참 |
| `[1, 2, 3]` | **True** | 요소가 있으면 참 |

```python
print(bool(1))
print(bool('hello'))
print(bool(0))
print(bool(''))
print(bool([]))
```
