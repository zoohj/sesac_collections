### 문제 1
나이를 입력받아 다음 조건에 해당하면 "무료 입장입니다"를 출력하고, 그렇지 않으면 "유료 입장입니다"를 출력하세요.
- 조건: 나이가 7세 이하이거나 65세 이상인 경우

```python
age = int(input())
if age <= 7 or age >= 65:
    print("무료")
else:
    print("유료")
```

### 문제 4
아이디(`user_id`)와 비밀번호(`user_pw`)를 입력받습니다.
- 아이디가 `"admin"`이고 비밀번호가 `"1234"`인 경우: "로그인 성공" 출력
- 아이디가 `"admin"`이지만 비밀번호가 틀린 경우: "비밀번호가 틀렸습니다" 출력
- 아이디가 틀린 경우: "존재하지 않는 아이디입니다" 출력

```python
user_id, user_pw = input().split()
if user_id == "admin":
    if user_pw == "1234":
        print("로그인 성공")
    else:
        print("비밀번호가 틀렸습니다.")
else:
    print("존재하지 않는 아이디입니다")
```


### 문제 7
연도(`year`)를 입력받아 윤년인지 아닌지 판별하는 프로그램을 작성하세요.
- 조건 1: 연도가 4로 나누어 떨어지면 윤년입니다.
- 조건 2: 하지만 그중 100으로 나누어 떨어지는 연도는 윤년이 아닙니다.
- 조건 3: 단, 400으로 나누어 떨어지는 연도는 여전히 윤년입니다.

결과로 "윤년입니다" 또는 "평년입니다"를 출력하세요.

```python
year = int(input())
if (year % 4 == 0):
    if year % 100 == 0:
        if year % 400 == 0:
            print("윤년")
        else:
            print("평년")
    else:
        print("윤년")
```

### 문제 8
영화관 입장료는 기본 12,000원입니다. 다음 조건에 따라 최종 요금을 계산하여 출력하세요.
- 나이가 13세 미만이면 50% 할인
- 나이가 65세 이상이면 70% 할인
- '쿠폰'이 있는 경우 추가로 2,000원 할인

최종적으로 "당신의 입장료는 OO원입니다."라고 출력하는 코드를 작성하세요.

```python
cost = 12000
age = int(input())
coupon = bool(input( "쿠폰이 있으면 1, 없으면 0 입력: "))
if age < 13:
    cost_before_coupon= cost * 0.5
elif age >= 65:
    cost_before_coupon= cost * 0.7
else:
    cost_before_coupon= cost
if coupon:
    cost_after_coupon= cost_before_coupon - 2000
    print(cost_after_coupon)
```
