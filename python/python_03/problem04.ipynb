
### 문제 1
사용자로부터 정수 `n`을 입력받아, 1부터 `n`까지 모든 정수의 합을 출력하세요.

````python
n= int(input())

sum=0
for i in range(1,n+1):
    print(i)
    sum += i
    
print(sum)

# while True:
#     print(num)
#     num +=1
#     if num > n:
#         break
````

### 문제 2

지정된 비밀번호(예: `"python123"`)가 입력될 때까지 계속해서 입력을 받는 프로그램을 작성하세요. 틀릴 경우 "다시 시도하세요"를 출력합니다.

````python
correctpw="python123"
trypw=""
while trypw!=correctpw:
    trypw=input("비밀번호 입력")
    if trypw==correctpw:
        print("굿굿")
    else:
        print("다시 시도ㄱㄱ")        
````

### 문제 3

1부터 30까지의 숫자 중에서 3의 배수를 제외한 나머지 숫자들만 출력하세요.

````python
start=1
end=30
for i in range(start,end):
    if not i%3==0:
        print(i)
````

### 문제 4
`names = ['jun', 'ken', 'alex', 'sophia', 'neo']` 일 때,

출석번호를 포함하여 '1번 jun', ..., '5번 neo'의 형태로 출석을 부르시오.

````python
names = ['jun', 'ken', 'alex', 'sophia', 'neo']
for i in range(len(names)):
    print(i+1,"번",names[i],end=" ,")
````

### 문제 6

사용자로부터 숫자를 계속 입력받다가, `0`을 입력하면 지금까지 입력된 모든 숫자의 합계를 출력하고 종료하는 프로그램을 작성하세요.

````python
sum=0
while True:
    num= int(input())
    if num != 0:
        sum += num
    else:
        print(sum)
        break
    
````

### 문제 7

1부터 20까지의 숫자 중 홀수는 건너뛰고, 짝수들만 더하다가 합계가 40을 넘으면 멈추고 합계를 출력하는 프로그램을 작성하세요.

````python
total_sum=0
start=1
end=20

for i in range(start,end+1):
    if i % 2 == 0:
        total_sum += i
        if total_sum > 40:
            print(total_sum)
            break
````
