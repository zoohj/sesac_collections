### 문제 6

### 문제 7
숫자 두개를 각각 입력받아 두 수 사이에 속한 모든 정수의 합을 구하시오. (ex. 2와 5를 입력받는다면 2 + 3 + 4 + 5 = 14)

```python
s, e = map(int, input().split())
result=0

while (s<=e):
    result +=s
    s+=1

print(result)
```

### 문제 8
다음 리스트 `colors = ['red', 'green', 'blue', 'yellow', 'purple']`의 인덱스와 값을 함께 출력하세요.
(출력 예시: `번호: 0, 색상: red` ...)

```python
colors = ['red', 'green', 'blue', 'yellow', 'purple']

# range(len(colors))를 활용하여 작성해보세요.
for i in range(len(colors)):
    color = colors[i]
    print(f"번호: {i}, 색상: {color}")
```

### 문제 9
1부터 50까지의 숫자 중에서 5의 배수만 골라 `my_list`에 저장하고, 그 리스트의 모든 요소의 합을 `for`문을 사용하여 구하세요.

```python
my_list = []
total = 0

# 1. range를 사용하여 5의 배수를 리스트에 추가
# 2. for문을 사용하여 합계 계산

for i in range(5,51,5):
    my_list.append(i)

for i in my_list:
    total+=i


print(f"리스트: {my_list}")
print(f"합계: {total}")
```

### 문제 11
사용자로부터 숫자를 계속 입력받다가 `-1`을 입력하면 입력을 멈추고, 그동안 입력받은 숫자들 중에서 **가장 큰 값**을 출력하는 프로그램을 `while`문을 사용하여 작성하세요. (단, `max()` 함수를 사용하지 마세요.)

```python
max_val = 0

while True:
    num = int(input("숫자를 입력하세요 (-1 종료): "))
    if num == -1:
        break
    # 여기에 최댓값을 갱신하는 로직을 작성하세요
    if num > max_val:
        max_val=num

print(f"가장 큰 값: {max_val}")
```

### 문제 12
`while`문을 사용하여 1부터 100까지의 정수 중 홀수의 합과 짝수의 합을 각각 따로 구하여 출력하세요.

```python
odd_sum = 0
even_sum = 0
n = 1

while n <= 100:
    
    # 조건문을 사용하여 홀수/짝수 합계를 구하세요
    if n%2==0:
        even_sum+=n
    else:
        odd_sum+=n
        
    n+=1

print(f"홀수의 합: {odd_sum}")
print(f"짝수의 합: {even_sum}")
```

### 문제 13
