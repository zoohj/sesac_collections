### 문제 2
구구단 3단을 출력하시오. 단, for, while을 활용하여 다양하게 만들어보시오.

```python
# [3, 6, 9, 12, ..., 27]
n = 3
m = 9

print([num for num in range(n, n*m + 1, n)])

lst = []
for i in range(1, m + 1):
    lst.append(n * i)
print(lst)

lst = [n]
for _ in range(m - 1):
    lst.append(lst[-1] + n)
print(lst)
```

### 문제 7

```python
n = int(input())
m = int(input())

# n은 작은 것 / m은 큰 것
if n > m:
    n, m = m, n # swap

total = 0
for num in range(n, m+1):
    total += num
print(total)
    

print((n + m) * (m - n + 1)//2) # float이기 때문에 int로 바꿔준다.
print(int((n + m) * (m - n + 1)/2)) # float이기 때문에 int로 바꿔준다.
```

### 문제 11
사용자로부터 숫자를 계속 입력받다가 `-1`을 입력하면 입력을 멈추고, 그동안 입력받은 숫자들 중에서 **가장 큰 값**을 출력하는 프로그램을 `while`문을 사용하여 작성하세요. (단, `max()` 함수를 사용하지 마세요.)

```python
# 숫자를 계속 입력받다가 -1이면 입력을 멈추자.

# 숫자들의 큰 값을 저장하자.
max_value = float('-inf')
while True:
    num = int(input())

    if num > max_value:
        max_value = num
    
    if num == -1:
        break
print(max_value)
```

```python
# 숫자를 계속 입력받다가 -1이면 입력을 멈추자.

# 숫자들의 큰 값을 저장하자.
num = int(input())
max_value = num
while num != -1:
    num = int(input())
    if num > max_value:
        max_value = num

print(max_value)
```

### 문제 12
`while`문을 사용하여 1부터 100까지의 정수 중 홀수의 합과 짝수의 합을 각각 따로 구하여 출력하세요.

```python
# 1. 1 ~ 100까지의 숫자를 하나하나 확인함.
# 2. 홀수 : 홀수로 더하기
#    짝수 : 짝수로 더하기

n = 1
total_even = 0
total_odd = 0
while n <= 100:

    if n%2 == 0: # 짝수
        total_even += n
    else: # 홀수
        total_odd += n

    n += 1
print(total_even, total_odd)
```

### 문제 14

(추가로, n을 입력받아 같은 과정을 반복하시오.)

```python
# 소수는 무엇이야?
# 1이랑 자기자신으로 나누었을 때"만" 나머지가 0인 수.
    # 위의 조건에서 "만"을 빼면 사실은 모든 수가 만족합니다

# 1이랑 자기자신으로 나누었을 때 나머지가 0이다
# 2 ~ 자기자신 -1로 나누었을 때 나머지가 0이 아니다.
    # 모든 경우에 대해서 싹다 0이 아니여야만 소수입니다.

# 이 경우는 모든 경우에 대해서 확인을 해야 결과가 나옴.
###########################
# 소수를 구하는 것이 아니라, 소수가 아님을 구하자.
# 2 ~ 자기자신 -1 로 나누었을 때 나머지가 0인 것이 존재하면 소수가 아니다.

# 특정한 경우에만 확인이 되면 결과가 나옴.

# -> 후자를 사용할 예정이다.
# 즉, 2 ~ 자기자신 -1까지 확인해서
# 나머지가 0이면 stop. 너는 소수가 아니야.
# 그러기 위해서는 초기값 -> 소수가 맞다로 잡는겁니다.


```

```python
# remove

numbers = [1, 2, 3, 4, 2]
# numbers.remove(4)
# print(numbers)

removal_el = 2
is_removed = False # flag

# 고대로 받아적을꺼야. 단, 특정 el면 안적을꺼야. 근데 1번만 안적을꺼야.
lst = []
for el in numbers:
    # 특정 원소이고 / 아직 remove 안했으면
    if el == removal_el and not is_removed:
        is_removed = True
        continue
    lst.append(el)
print(lst)
```
