### 문제 1
API로 전달받은 게시글 데이터입니다. `status`가 'published'인 게시글들만 필터링하여, 해당 게시글들의 `view_count` 총합을 구하세요.

```python
posts = [
    {'title': '첫 번째 글', 'status': 'published', 'view_count': 120},
    {'title': '두 번째 글', 'status': 'draft', 'view_count': 0},
    {'title': '세 번째 글', 'status': 'published', 'view_count': 350},
    {'title': '네 번째 글', 'status': 'deleted', 'view_count': 10}
]

# print(type(posts))
# print(type(posts[0]))

# 1. 각 게시글의 status를 확인한다.
# 2. 걔가 published인지를 확인한다.
# 3. 그것들의 view count를 확인하고, 합을 구한다.

# post = posts[0]
# print(post['status'])

total_view_count = 0
for post in posts:
    status = post['status']
    if status == 'published':
        view_count = post['view_count']
        total_view_count += view_count
        # print(post)
        # print()
print(total_view_count)
```
