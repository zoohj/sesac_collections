### 문제 1
입력받은 점수에 따라 합격/불합격을 판별하고 메시지를 반환하는 함수 `get_grade_result`를 작성하세요.

**조건:**
1. 점수가 60점 이상이면 "합격"을 반환합니다.
2. 점수가 60점 미만이면 "불합격"을 반환합니다.
3. 만약 점수가 0점보다 작거나 100점보다 크면 "잘못된 점수"를 반환하세요.

```python
def get_grade_result(score):
    # 코드를 작성하세요
    if score >= 60 and score <=100:
        return "합격"
    elif score >= 0 and score <60:
        return "불합격"
    else:
        return "잘못된 점수"

# 테스트
print(get_grade_result(85))  # 결과: "합격"
print(get_grade_result(40))  # 결과: "불합격"
print(get_grade_result(120)) # 결과: "잘못된 점수"
```

### 문제 2
입력받은 숫자가 홀수인지 짝수인지 판별하여 문자열을 반환하는 함수 `check_even_odd`를 작성하세요.

**조건:**
1. 숫자가 짝수이면 "짝수"를 반환합니다.
2. 숫자가 홀수이면 "홀수"를 반환합니다.

```python
def check_even_odd(number):
    check= number % 2
    if check == 0:
        return "짝수"
    else:
        return "홀수"

# 테스트
print(check_even_odd(10)) # 결과: "짝수"
print(check_even_odd(7))  # 결과: "홀수"
```

### 문제 3
두 개의 숫자와 연산자(+, -)를 입력받아 계산 결과를 반환하는 아주 간단한 계산기 함수 `simple_calc`를 작성하세요.

**조건:**
1. `op`가 "+"이면 두 수의 합을 반환합니다.
2. `op`가 "-"이면 두 수의 차를 반환합니다.
3. 그 외의 연산자가 들어오면 "지원하지 않는 연산"이라는 메시지를 반환하세요.

```python
def simple_calc(num1, num2, op):
    # 코드를 작성하세요
    if op == "+":
        return num1 + num2 
    elif op == "-":
        return num1 - num2 
    else:
        return "지원하지 않는 연산"

# 테스트
print(simple_calc(10, 5, "+")) # 결과: 15
print(simple_calc(10, 5, "-")) # 결과: 5
print(simple_calc(10, 5, "*")) # 결과: "지원하지 않는 연산"
```

### 문제 4
이름 리스트를 입력받아 이름이 3자 이상인 사람만 골라 새로운 리스트로 반환하는 함수 `filter_long_names`를 작성하세요.

**조건:**
1. `for`문을 사용하여 리스트의 각 이름을 확인합니다.
2. `len()` 함수를 사용하여 이름의 길이를 확인합니다.

```python
def filter_long_names(names):
    # 코드를 작성하세요
    long_names=[]
    for name in names:
        if len(name)>=3:
            long_names.append(name)
    return long_names

# 테스트
name_list = ["김철수", "이영희", "박씨", "최도날드"]
print(filter_long_names(name_list)) # 결과: ["김철수", "이영희", "최도날드"]
```

### 문제 5
회원가입 서비스에서 유저가 입력한 데이터의 유효성을 검사하는 함수 `validate_user_registration`을 작성하세요.

**조건:**
1. `username`은 4자 이상이어야 합니다.
2. `age`는 19세 이상이어야 합니다.
3. `email`에는 반드시 '@' 기호가 포함되어야 합니다.
4. 위 조건을 모두 만족하면 `True`, 하나라도 실패하면 `False`를 반환합니다.

```python
def validate_user_registration(username, age, email):
    if len(username) >= 4 and age >= 19 and '@' in email:
        return True
    else:
        return False

# 테스트
print(validate_user_registration("kimpython", 25, "test@example.com")) # 결과: True
print(validate_user_registration("kim", 25, "test@example.com"))      # 결과: False
```

### 문제 6
장바구니에 담긴 상품들의 최종 결제 금액을 계산하는 함수 `calculate_checkout_price`를 작성하세요.

**조건:**
1. `membership_grade`가 'VIP'이면 총액의 15%, 'GOLD'이면 10%를 할인합니다.
2. 할인 적용 후 최종 금액이 50,000원 미만이면 배송비 3,000원을 추가합니다.
3. 50,000원 이상이면 배송비는 무료(0원)입니다.

```python
def calculate_checkout_price(cart_items, membership_grade):
    # cart_items는 {'price': 정수, 'quantity': 정수} 형태의 딕셔너리 리스트입니다.
    # 코드를 작성하세요
    sum_before_sale=0
    sum_after_sale=0
    total_count=0
    for i in range(len(cart_items)):
        item = cart_items[i]
        sum_before_sale += item["price"] * item["quantity"]

    # 총액 15퍼 할인
    if membership_grade == "VIP":
        sum_after_sale = sum_before_sale * 0.85

    # 총액 10퍼 할인
    elif membership_grade == "GOLD":
        sum_after_sale = sum_before_sale * 0.90


    if sum_after_sale < 50000:
        total_count = sum_after_sale + 3000
    else:
        total_count = sum_after_sale

    return int(total_count)

my_cart = [{"price": 25000, "quantity": 1}, {"price": 15000, "quantity": 2}]
print(calculate_checkout_price(my_cart, "GOLD")) # 결과: 52500
```

### 문제 7
서버에서 받아온 다양한 형태의 데이터가 섞인 리스트에서 정수 ID만 추출하여 정형화하는 함수 `normalize_user_data`를 작성하세요.

**조건:**
1. 숫자형 문자열("105"), 숫자(101), 리스트 안의 숫자([102]) 등을 모두 정수로 변환하세요.
2. 중복된 ID는 제거하고, 결과는 오름차순으로 정렬된 리스트여야 합니다.

```python
def normalize_user_data(raw_data):
    clean_data=[]
    for data in raw_data:
        #문자열, 숫자, 리스트 안의 숫자
        # if type(data) == 'str':
        if isinstance(data, str):
            clean_data.append(int(data))
        elif isinstance(data, int):
            clean_data.append(data)
        elif isinstance(data, list):
            for i in range(len(data)):
                clean_data.append(data[i])
    clean_data=list(set(clean_data))
    clean_data.sort() # => 리스트가 되나?
    return clean_data

dirty_data = ["105", 101, [102, 103], "101", 104, [105]]
print(normalize_user_data(dirty_data)) # 결과: [101, 102, 103, 104, 105]
```

### 문제 8
특정 조건을 만족하는 유저들을 추출하여 알림 대상자 리스트를 반환하는 함수 `get_notification_targets`를 작성하세요.

**조건:**
1. `is_active`가 `True`인 유저 중에서 선별합니다.
2. `last_login`이 7일 이상이거나, `point`가 10,000점 이상인 유저가 대상입니다.
3. 대상자의 'name'만 리스트에 담아 반환하세요.

```python
def get_notification_targets(users_dict):
    noti_user=[]
# 알림 대상자의 이름 리스트 반환, is_active=True 반환, last_login >= 7 or point >= 10000
    for user in users_dict.values():
        if user['is_active']==True:
            if user['last_login'] >= 7 or user['point'] >=10000:
                noti_user.append(user['name'])
    return noti_user


users_data = {
    "u1": {"name": "Alice", "is_active": True, "last_login": 10, "point": 5000},
    "u2": {"name": "Bob", "is_active": False, "last_login": 1, "point": 15000},
    "u3": {"name": "Charlie", "is_active": True, "last_login": 2, "point": 12000}
}
print(get_notification_targets(users_data)) # 결과: ['Alice', 'Charlie']
```
