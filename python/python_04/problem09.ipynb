### 문제 1
여러 명의 학생 성적 데이터가 담긴 복합 리스트입니다. 각 학생의 `grades` 리스트에서 80점 이상의 점수만 추출하여 평균을 계산하고, 그 평균이 90점 이상인 학생의 이름만 `top_students` 리스트에 담으세요.

```python
students = [
    {"name": "John", "grades": [85, 92, 78, 100]},
    {"name": "Sarah", "grades": [88, 95, 90, 89]},
    {"name": "Tyler", "grades": [70, 75, 82, 80]},
    {"name": "Emma", "grades": [95, 98, 88, 92]}
]

top_students=[]
for student in students:
    good_score=[]
    count=0
    for grade in student["grades"]:
        if grade >= 80:
            count+=1
            good_score.append(grade)
    avg = sum(good_score)/count
    if avg >= 90:
        top_students.append(student["name"])

print(top_students)
    
```

### 문제 2
소셜 미디어 API 응답 데이터 형태입니다. 각 유저(`users`)가 작성한 게시글(`posts`)들 중, 좋아요(`likes`) 수가 50개 이상인 게시글의 제목(`title`)만 모두 모아 하나의 리스트로 만드세요.

```python
api_response = {
    "status": "success",
    "data": {
        "users": [
            {
                "username": "user1",
                "posts": [
                    {"title": "오늘의 식단", "likes": 30},
                    {"title": "여행 사진", "likes": 150}
                ]
            },
            {
                "username": "user2",
                "posts": [
                    {"title": "코딩 공부법", "likes": 88},
                    {"title": "운동 일지", "likes": 10}
                ]
            },
            {
                "username": "user3",
                "posts": [
                    {"title": "ai 공부법", "likes": 808},
                    {"title": "일기", "likes": 100}
                ]
            },
        ]
    }
}

popular_titles = []
users= api_response["data"]["users"]
for user in users:
    posts = user["posts"]
    for post in posts:
        if post["likes"] >= 50:
            popular_titles.append(post["title"])


# 코드를 작성하세요

print(f"인기 게시글 제목: {popular_titles}")
```

### 문제 3
상품 리스트를 카테고리별로 분류하여 딕셔너리를 만드세요.

```python
products = [
    {"category": "전자제품", "name": "키보드"},
    {"category": "의류", "name": "티셔츠"},
    {"category": "전자제품", "name": "마우스"},
    {"category": "전자제품", "name": "노트북"},
    {"category": "식품", "name": "사과"},
    {"category": "식품", "name": "배"},
    {"category": "의류", "name": "청바지"}
]

grouped_data = {}
# grouped_data = { "category": ["name1", "name2", ...] }

for product in products:
    category = product["category"]
    if category not in grouped_data:
        grouped_data[category]=[]
    grouped_data[category].append(product["name"])

print(f"그룹화 결과: {grouped_data}")
```
