## 1. 문자열(String)
### 1-1. 문자열이란?
- 문자열은 0개 이상의 문자를 순서 있게 저장하는 컨테이너 자료형이다.
- 작은 따옴표(') 혹은 큰 따옴표(")를 이용해 생성한다.
- 문자열은 연속으로 나열된 문자들의 집합체라고 할 수 있다.
  - ex) "python" 문자열은 "p", "y", "t", "h", "o", "n" 의 6개의 문자들이 연속으로 나열된 집합체이다.
- 빈 문자열도 문자열이고, 원소가 1개인 문자열도 문자열이다. 또한 공백도 문자열이다.

```python
string = "python"
empty = ""  # 빈 문자열
one = "p"
space = " "  # 공백
string = 'python'
"""
이렇게
줄바꿈을 해도
유지되는
문자열을 만들 수 있어요.
"""


# 내장함수 len을 사용하면 문자열의 길이를 알 수 있다.
sentence = "hello python"
print(len(sentence))  # sentence의 길이 출력 (공백 포함 12글자)
```

- 내장함수 str을 사용하면 다른 컨테이너 자료형을 문자열로 형변환할 수 있다.

```python
number_list = [10, 20, 30, 40, 50]
number_list_string = str(number_list)  # 리스트를 문자열로 형변환

print(number_list_string)
print(type(number_list_string))
```

### 1-2. 문자열 연산
문자열은 덧셈(+)과 곱셈(*) 연산을 제공한다.

#### 문자열 덧셈(+)
두 개 이상의 문자열을 합칠 수 있다.

```python
a = "hello"
b = "python"
c = a + b

print(c)
```

#### 문자열 곱셈(*)
문자열의 원소를 여러 번 반복하여 생성할 수 있다.

```python
a = "안녕"
b = a * 3

print(b)
```

### 1-3. 인덱싱(Indexing)
#### 접근(조회)
문자열[인덱스]의 형식으로 원소에 접근할 수 있다.

```python
word = "python"
print(word[1])  # 문자열 word의 1번째 원소인 "y"를 출력
```

- 문자열도 리스트와 튜플처럼 음수 인덱싱이 가능하다.

```python
word = "python"
print(word[-1])  # 문자열 word의 -1번째(마지막) 원소인 "n"을 출력
```

#### 문자열은 수정이 불가능
- 문자열은 변경 불가능한(immutable) 컨테이너이므로, 인덱싱을 이용해 수정할 수 없다.
- 아래 코드는 에러를 발생시킨다.

```python

word = "python"
word[0] = "j"
```

“python”에서 “p”를 “j”로 바꾸고 싶다면, 덧셈과 슬라이싱을 이용해 새로운 문자열을 생성해야 한다.

```python
word = "python"
new_word = "j" + word[1:]  # 덧셈과 슬라이싱을 이용하여 마치 수정한 것처럼 보임

print(new_word)
print("이는 새로운 문자열을 생성한 것이지, 기존의 문자열을 수정한 것은 아닙니다.")

word[:2] + 'haha' + word[2:]
```

### 1-4. 슬라이싱(Slicing)
- 문자열도 리스트와 튜플처럼 슬라이싱이 가능하다.
- `문자열[start:end:step]`의 형식으로 작성한다.
  - **start**: 시작 인덱스
  - **end**: 끝 인덱스 (포함되지 않음)
  - **step**: 간격 (음수는 반대 방향)

```python
sentence = "abcdefghi"

# start ~ end-1 까지 자르기 (양수, 음수 인덱스 혼합 가능)
print(sentence[2:5])     # cde
print(sentence[-6:-2])   # defg
print(sentence[2:-4])    # cde

# step을 이용해서 자르기
print(sentence[2:5:2])   # ce
print(sentence[-6:-1:3]) # dg
print(sentence[2:5:-1])  # 빈 문자열
print(sentence[5:2:-1])  # fed

# 기본값으로 자르기
print(sentence[:3])      # abc
print(sentence[5:])      # fghi
print(sentence[:])       # abcdefghi
print(sentence[::-1])    # ihgfedcba (문자열 뒤집기)
```

## 2. 문자열의 멤버십 연산
### 2-1. in 연산자
해당 원소가 문자열에 있으면 `True`, 없으면 `False`를 반환한다.

```python
sentence = "hello"

print("h" in sentence)  # True
print("j" in sentence)  # False
```

### 2-2. not in 연산자
해당 원소가 문자열에 없으면 `True`, 있으면 `False`를 반환한다.

```python
sentence = "hello"

print("h" not in sentence)  # False
print("j" not in sentence)  # True
```

## 3. 문자열의 반복
### 3-1. range와 len을 이용한 반복

```python
sentence = "hello"

for i in range(len(sentence)):
    print(sentence[i])
```

### 3-2. 문자열 직접 반복

```python
sentence = "hello"

for char in sentence:
    print(char)
```

### 3-3. enumerate를 이용한 반복

```python
sentence = "hello"

for i, char in enumerate(sentence):
    print(f"index: {i}, char: {char}")
```

## 4. 문자열 포매팅
### 4-1. 덧셈을 이용한 방법

```python
name = "kyle"
sentence = "제 이름은 " + name + " 입니다."
print(sentence)
```

### 4-2. f-string을 이용한 방법
- 따옴표 앞에 `f`를 붙인다.
- 변수나 식은 중괄호(`{}`)로 감싼다.

```python
# 3 X 1 = 3
# 3 X 2 = 6
# 3 X 3 = 9
for i in range(1, 4):
    # print("3 X ", i, "=", i*3)
    print(f"3 X {i} = {i*3}")
```

```python
name = "kyle"
age = 20
sentence = f"{name}은 {age}살 입니다."
print(sentence)

# 연산 및 함수 호출 포함
print(f"이름의 길이는 {len(name)} 입니다.")
print(f"내년이 되면 {age + 1}살 입니다.")
```

```python
# .format을 활용한 문자열 포매팅
# 위의 f-string의 경우 - 알고리즘 및 기타 다양한 상황.
# .format은 예전 문법이지만, prompting을 할 때 더 많이 사용됨.

sentence = "{subject}에 대해서 자세히 알려줘{subject}."
print(sentence.format(subject='아이유'))

print(sentence.format(subject='윈터'))
```

## 5. 자주 사용하는 문자열 메서드
### 5-1. split
구분자를 기준으로 문자열을 나누어 리스트로 반환한다.

```python


sentence = "hello$python$!"
print(sentence.split("$"))

sentence2 = "hello python !"
print(sentence2.split())  # 공백 기준
```

### 5-2. strip, lstrip, rstrip
특정 문자(기본값은 공백/개행)를 제거한다.

```python
sentence = "$hello$"
print(sentence.strip("$"))

sentence2 = " hello "
print(sentence2.strip())

sentence3 = "hello"
print(sentence3.lstrip("h"))
print(sentence3.rstrip("o"))
```

### 5-3. find, index
특정 문자열의 인덱스를 찾는다.

```python
sentence = "hello"
print(sentence.find("e"))   # 1
print(sentence.index("e"))  # 1

print(sentence.find("j"))   # 존재하지 않으면 -1
# print(sentence.index("j")) # 존재하지 않으면 에러 발생
```

### 5-4. count
특정 문자열의 갯수를 반환한다.

```python
sentence = "hello"
print(sentence.count("l"))
```

### 5-5. replace
문자열을 변경한다. (원본은 유지되고 변경된 새로운 문자열 반환)

```python
sentence = "hello"
print(sentence.replace("h", "j"))
print(sentence.replace("l", "x"))

# 삭제 용도로 사용
print(sentence.replace("h", ""))
```

### 5-6. join
컨테이너의 원소들을 특정 구분자로 이어붙인다.

```python
string_list = ["hello", "python", "!"]
print(" ".join(string_list))
print("".join(["h", "e", "l", "l", "o"]))
print("\n".join(string_list))
```

## 6. 아스키코드
### 6-1. 비트와 바이트
- **bit(비트)**: 0과 1을 나타내는 최소 단위
- **byte(바이트)**: 8비트를 모은 단위 (256가지 표현 가능, 데이터 저장 기본 단위)

### 6-2. 아스키코드(ASCII)란?
- 영문 알파벳을 숫자로 표현하는 표준 규칙.
- 각 문자는 1 byte를 사용한다.

### 6-3. 내장함수 ord, chr

```python
# ord(문자) : 문자를 아스키코드로 변환
print(ord("A"))
print(ord("a"))

# chr(숫자) : 아스키코드를 문자로 변환
print(chr(65))
print(chr(97))
```
