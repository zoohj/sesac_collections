# 파이썬 클래스 상속 연습 문제

### 문제 1
동물을 나타내는 `Animal` 클래스와 이를 상속받는 `Dog`, `Cat` 클래스를 작성하세요.
1. `Animal` 클래스는 생성자(`__init__`)에서 이름(`name`)을 저장하고, `move(self)` 메서드에서 "[이름]이 이동합니다."를 출력합니다.
2. `Dog` 클래스는 `Animal`을 상속받으며, 고유 메서드 `bark(self)`에서 "[이름]이 멍멍 짖습니다."를 출력합니다.
3. `Dog` 클래스는 `Animal`을 상속받으며, 고유 메서드 `mew(self)`에서 "[이름]이 냥냥 웁니다."를 출력합니다.
4. `Dog`, `Cat` 인스턴스를 생성하고 두 메서드를 모두 호출해보세요.

```python
# 여기에 코드를 작성하세요.
class Animal:
    def __init__(self, name):
        self.name= name
    def move(self):
        return print(f"{self.name}이 이동합니다")

class Dog(Animal):
    def bark(self):
        return print(f"{self.name}이 멍멍 짖습니다")

class Cat(Animal):
    def mew(self):
        return print(f"{self.name}이 냥냥 웁니다")
dog = Dog("진돌")
dog.bark()
cat = Cat("나비")
cat.mew()

dog.move()
```

### 문제 2
전자제품을 나타내는 `Electronics` 클래스와 이를 상속받는 `Smartphone`, `Fan` 클래스를 작성하세요.
1. `Electronics` 클래스는 `brand`를 속성으로 가지며, `power_on(self)` 메서드에서 "[brand] 제품의 전원이 켜집니다."를 출력합니다.
2. `Smartphone` 클래스는 `super()`를 사용하여 부모의 `brand`를 초기화하고, 추가로 모델명(`model`)을 속성으로 가집니다.
3. `Smartphone` 클래스에서 `power_on(self)` 메서드를 오버라이딩하여, 부모의 기능을 먼저 실행한 후 "모델 [model]이 부팅을 시작합니다."를 추가로 출력하세요.
4. `Fan` 클래스에서 `power_on(self)` 메서드를 오버라이딩하여, 부모의 기능을 먼저 실행한 후 "바람이 붑니다"를 추가로 출력하세요.

```python
# 여기에 코드를 작성하세요.

class Electronics:
    def __init__(self, brand):
        self.brand = brand
    def power_on(self):
        return print(f"{self.brand} 제품의 전원이 켜집니다.")

class Smartphone(Electronics):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model
    def power_on(self):
        super().power_on()
        print(f"모델 {self.model}이 부팅을 시작합니다.")

class Fan(Electronics):
    def power_on(self):
        super().power_on()
        print("바람이 붑니다")
```

### 문제 3
두 개의 부모 클래스로부터 기능을 물려받는 자식 클래스를 작성하세요.
1. `Speaker` 클래스: `play_sound(self)` 메서드에서 "소리를 출력합니다."를 출력합니다.
2. `Camera` 클래스: `take_photo(self)` 메서드에서 "사진을 촬영합니다."를 출력합니다.
3. `SmartDevice` 클래스: `Speaker`와 `Camera`를 모두 상속받습니다.
4. `SmartDevice` 인스턴스를 생성하여 두 부모 클래스의 메서드를 모두 호출해보세요.

```python
# 여기에 코드를 작성하세요.
class Speaker:
    def play_sound(self):
        return print("소리를 출력합니다")

class Camera:
    def take_photo(self):
        print("사진을 촬영")

class SmartDevice(Speaker, Camera):
    def play_sound(self):
        return super().play_sound()
    def take_photo(self):
        return super().take_photo()
    

new = SmartDevice()

new.play_sound()
new.take_photo()
```
